Importing the necessary modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import io
```

Loading the dataset

```python
uploaded = files.upload()
df = pd.read_csv(io.BytesIO(uploaded['user_timings final.csv']))
df
```

Checking for any redundant values in the dataset

```python
df.duplicated()
```

Dividing the data into features and labels

```python
X = df.drop(['subject', 'sessionIndex', 'rep'], axis=1)
y = df['subject']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from keras.models import Sequential
from keras.layers import Dense
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
y_encoded
```

Dividing the data into training and testing part

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.4, random_state=42)
```

Standardize features using StandardScaler

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Building a simple neural network model with keras

```python
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(len(label_encoder.classes_), activation='softmax'))
```

Compiling the model

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

Training the model

```python
model.fit(X_train_scaled, y_train, epochs=10, batch_size=32)
```

Evaluate the model on the test set

```python
accuracy = model.evaluate(X_test_scaled, y_test)[1]
print(f'Model Accuracy: {accuracy}')
```
